```python
üìå Exerc√≠cio TP1: An√°lise de Compras e Gest√£o de Funcion√°rios do Governo com PySpark üöî
O Governo adquiriu ve√≠culos para v√°rias entidades p√∫blicas e precisa de analisar n√£o s√≥ as compras, mas tamb√©m a gest√£o dos funcion√°rios respons√°veis por conduzir ou manter os ve√≠culos.
1Ô∏è‚É£ Criar a base de dados e as tabelas
O teu objetivo √© criar duas tabelas relacionadas dentro da base de dados compras_governo:
üìä Tabela 1: compras_veiculos (Registo das compras de ve√≠culos)
Nome da Coluna
Tipo de Dados
Descri√ß√£o
id_compra
INT
Identificador √∫nico da compra
entidade
STRING
Nome da entidade governamental
marca
STRING
Marca do ve√≠culo comprado
modelo
STRING
Modelo do ve√≠culo
ano
INT
Ano de fabrico
preco
FLOAT
Pre√ßo unit√°rio do ve√≠culo (‚Ç¨)
quantidade
INT
N√∫mero de unidades compradas
fonte_financiamento
STRING
Fonte de financiamento utilizada

üîπ Exemplo de dados:
id_compra
entidade
marca
modelo
ano
preco
quantidade
fonte_financiamento
1
Pol√≠cia Nacional
BMW
S√©rie 5
2023
55000.0
5
Or√ßamento do Estado
2
Bombeiros Municipais
Renault
Master
2022
35000.0
3
Fundo Europeu
3
Minist√©rio da Sa√∫de
Mercedes
Sprinter
2023
45000.0
8
Parceria P√∫blico-Privada


üìä Tabela 2: funcionarios (Gest√£o dos funcion√°rios respons√°veis pelos ve√≠culos)
Nome da Coluna
Tipo de Dados
Descri√ß√£o
id_funcionario
INT
Identificador √∫nico do funcion√°rio
nome
STRING
Nome do funcion√°rio
cargo
STRING
Cargo do funcion√°rio
entidade
STRING
Entidade onde trabalha (liga√ß√£o com compras_veiculos)
veiculo_designado
STRING
Modelo do ve√≠culo designado para o funcion√°rio

üîπ Exemplo de dados:
id_funcionario
nome
cargo
entidade
veiculo_designado
101
Manuel Silva
Motorista
Pol√≠cia Nacional
BMW S√©rie 5
102
Sofia Mendes
Bombeira
Bombeiros Municipais
Renault Master
103
Jo√£o Ferreira
T√©cnico de Manuten√ß√£o
Minist√©rio da Sa√∫de
Mercedes Sprinter












2Ô∏è‚É£ Consultas sobre os dados
Agora, com base nas tabelas compras_veiculos e funcionarios, resolve os seguintes desafios utilizando SQL no PySpark.
a) Listar todas as compras feitas pelo Governo com ve√≠culos designados para motoristas espec√≠ficos.
üìù Dica: Faz um JOIN entre compras_veiculos e funcionarios usando a coluna entidade.
b) Contar quantos ve√≠culos foram comprados por cada entidade governamental.
üìù Dica: Usa GROUP BY para contar o n√∫mero total de ve√≠culos por entidade.
c) Listar os funcion√°rios que trabalham em entidades que adquiriram ve√≠culos acima de 50.000‚Ç¨.
üìù Dica: Junta as duas tabelas e filtra os dados pelo preco.
d) Calcular o valor total gasto por cada entidade governamental.
üìù Dica: Multiplica preco * quantidade e usa GROUP BY para agrupar por entidade.
e) Adicionar uma coluna que indica se a entidade comprou ve√≠culos el√©tricos.
üìù Dica: Usa CASE WHEN para verificar se a marca √© "Tesla".



üîπ N√≠vel 1: Consultas Simples
1Ô∏è‚É£ Listar todas as compras feitas pelo Governo.
üìù Dica: Usa SELECT * FROM compras_veiculos.
2Ô∏è‚É£ Mostrar todas as compras financiadas pelo "Fundo Europeu".
üìù Dica: Filtra os resultados com WHERE fonte_financiamento = 'Fundo Europeu'.
3Ô∏è‚É£ Ordenar as compras pelo pre√ßo dos ve√≠culos em ordem decrescente.
üìù Dica: Usa ORDER BY preco DESC.
4Ô∏è‚É£ Contar quantos ve√≠culos foram comprados por cada entidade.
üìù Dica: Usa GROUP BY entidade e COUNT(*).
5Ô∏è‚É£ Calcular o valor total gasto por cada entidade governamental.
üìù Dica: Multiplica preco * quantidade e usa GROUP BY entidade.




üîπ N√≠vel 2: Consultas com Jun√ß√µes e Fun√ß√µes
6Ô∏è‚É£ Listar os funcion√°rios e os ve√≠culos que lhes foram designados.
üìù Dica: Faz um JOIN entre funcionarios e compras_veiculos usando entidade.
7Ô∏è‚É£ Contar quantos funcion√°rios est√£o alocados a cada entidade governamental.
üìù Dica: Usa COUNT(id_funcionario) com GROUP BY entidade.
8Ô∏è‚É£ Mostrar todas as compras feitas por entidades que t√™m mais de 5 funcion√°rios registados.
üìù Dica: Junta as tabelas e usa HAVING COUNT(id_funcionario) > 5.
9Ô∏è‚É£ Listar as compras feitas apenas por entidades que adquiriram ve√≠culos el√©tricos.
üìù Dica: Filtra pela marca = 'Tesla' ou outros ve√≠culos el√©tricos.
üîü Mostrar os funcion√°rios que utilizam ve√≠culos comprados ap√≥s 2022.
üìù Dica: Usa JOIN e filtra por ano > 2022.















üîπ N√≠vel 3: Consultas Mais Avan√ßadas
1Ô∏è‚É£1Ô∏è‚É£ Identificar entidades que gastaram mais do que 200.000‚Ç¨ em compras de ve√≠culos.
üìù Dica: Usa SUM(preco * quantidade) > 200000 com GROUP BY entidade.
1Ô∏è‚É£2Ô∏è‚É£ Determinar qual foi a compra mais cara feita pelo Governo e qual entidade a realizou.
üìù Dica: Usa MAX(preco * quantidade) e relaciona com a entidade correspondente.
1Ô∏è‚É£3Ô∏è‚É£ Listar todas as entidades que compraram ve√≠culos, mas n√£o t√™m funcion√°rios registados.
üìù Dica: Usa um LEFT JOIN e verifica valores NULL na tabela funcionarios.
1Ô∏è‚É£4Ô∏è‚É£ Para cada entidade, calcular a m√©dia de pre√ßos dos ve√≠culos adquiridos.
üìù Dica: Usa AVG(preco) GROUP BY entidade.
1Ô∏è‚É£5Ô∏è‚É£ Criar um ranking das entidades que mais gastaram com compras de ve√≠culos.
üìù Dica: Usa SUM(preco * quantidade), GROUP BY entidade e ORDER BY DESC para ranquear os gastos.

üìå Rela√ß√£o entre tabelas:
A tabela compras_veiculos regista as compras de ve√≠culos por diferentes entidades governamentais.
A tabela funcionarios cont√©m os funcion√°rios respons√°veis pelos ve√≠culos, e est√° relacionada com a tabela compras_veiculos atrav√©s do campo entidade.



            ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê
            ‚îÇ      compras_veiculos      ‚îÇ
            ‚îú‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î§
            ‚îÇ id_compra (PK)            ‚îÇ
            ‚îÇ entidade                   ‚îÇ‚óÑ‚îÄ‚îÄ‚îê
            ‚îÇ marca                      ‚îÇ   ‚îÇ
            ‚îÇ modelo                     ‚îÇ   ‚îÇ
            ‚îÇ ano                         ‚îÇ   ‚îÇ
            ‚îÇ preco                       ‚îÇ   ‚îÇ
            ‚îÇ quantidade                  ‚îÇ   ‚îÇ
            ‚îÇ fonte_financiamento         ‚îÇ   ‚îÇ
            ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò   ‚îÇ
                                            ‚îÇ  RELA√á√ÉO 1:N
            ‚îå‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îê   ‚îÇ
            ‚îÇ      funcionarios          ‚îÇ   ‚îÇ
            ‚îú‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î§   ‚îÇ
            ‚îÇ id_funcionario (PK)       ‚îÇ   ‚îÇ
            ‚îÇ nome                      ‚îÇ   ‚îÇ
            ‚îÇ cargo                     ‚îÇ   ‚îÇ
            ‚îÇ entidade                   ‚îÇ‚óÑ‚îÄ‚îÄ‚îò
            ‚îÇ veiculo_designado         ‚îÇ
            ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò






üìå Resolu√ß√£o üöî

 ‚Ä¶‚Ä¶. Resolu√ß√£o 
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, count, avg, max, when

# Criar sess√£o do Spark
spark = SparkSession.builder.appName("Compras_Governo").getOrCreate()

# Criar dados para compras_veiculos
dados_veiculos = [
    (1, "Pol√≠cia Nacional", "BMW", "S√©rie 5", 2023, 55000.0, 5, "Or√ßamento do Estado"),
    (2, "Bombeiros Municipais", "Renault", "Master", 2022, 35000.0, 3, "Fundo Europeu"),
    (3, "Minist√©rio da Sa√∫de", "Mercedes", "Sprinter", 2023, 45000.0, 8, "Parceria P√∫blico-Privada")
]
colunas_veiculos = ["id_compra", "entidade", "marca", "modelo", "ano", "preco", "quantidade", "fonte_financiamento"]
df_veiculos = spark.createDataFrame(dados_veiculos, colunas_veiculos)

dados_funcionarios = [
    (101, "Manuel Silva", "Motorista", "Pol√≠cia Nacional", "BMW S√©rie 5"),
    (102, "Sofia Mendes", "Bombeira", "Bombeiros Municipais", "Renault Master"),
    (103, "Jo√£o Ferreira", "T√©cnico de Manuten√ß√£o", "Minist√©rio da Sa√∫de", "Mercedes Sprinter")
]
colunas_funcionarios = ["id_funcionario", "nome", "cargo", "entidade", "veiculo_designado"]
df_funcionarios = spark.createDataFrame(dados_funcionarios, colunas_funcionarios)

# Consultas
print("1Ô∏è‚É£ Listar todas as compras feitas pelo Governo:")
df_veiculos.show()

print("2Ô∏è‚É£ Mostrar todas as compras financiadas pelo 'Fundo Europeu':")
df_veiculos.filter(col("fonte_financiamento") == "Fundo Europeu").show()

print("3Ô∏è‚É£ Ordenar as compras pelo pre√ßo dos ve√≠culos em ordem decrescente:")
df_veiculos.orderBy(col("preco").desc()).show()

print("4Ô∏è‚É£ Contar quantos ve√≠culos foram comprados por cada entidade:")
df_veiculos.groupBy("entidade").sum("quantidade").show()

print("5Ô∏è‚É£ Calcular o valor total gasto por cada entidade governamental:")
df_veiculos.groupBy("entidade").agg(sum(col("preco") * col("quantidade")).alias("total_gasto")).show()

print("6Ô∏è‚É£ Listar os funcion√°rios e os ve√≠culos que lhes foram designados:")
df_funcionarios.join(df_veiculos, "entidade").select("nome", "cargo", "entidade", "veiculo_designado").show()

print("7Ô∏è‚É£ Contar quantos funcion√°rios est√£o alocados a cada entidade governamental:")
df_funcionarios.groupBy("entidade").count().show()

print("8Ô∏è‚É£ Mostrar todas as compras feitas por entidades que t√™m mais de 5 funcion√°rios registados:")
df_funcionarios.groupBy("entidade").count().filter(col("count") > 5).join(df_veiculos, "entidade").show()

print("9Ô∏è‚É£ Listar as compras feitas apenas por entidades que adquiriram ve√≠culos el√©tricos:")
df_veiculos.filter(col("marca") == "Tesla").show()

print("üîü Mostrar os funcion√°rios que utilizam ve√≠culos comprados ap√≥s 2022:")
df_funcionarios.join(df_veiculos, "entidade").filter(col("ano") > 2022).select("nome", "cargo", "modelo").show()

print("1Ô∏è‚É£1Ô∏è‚É£ Identificar entidades que gastaram mais do que 200.000‚Ç¨ em compras de ve√≠culos:")
df_veiculos.groupBy("entidade").agg(sum(col("preco") * col("quantidade")).alias("total_gasto")).filter(col("total_gasto") > 200000).show()

print("1Ô∏è‚É£2Ô∏è‚É£ Determinar qual foi a compra mais cara feita pelo Governo e qual entidade a realizou:")
df_veiculos.withColumn("total", col("preco") * col("quantidade")).orderBy(col("total").desc()).limit(1).show()

print("1Ô∏è‚É£3Ô∏è‚É£ Listar todas as entidades que compraram ve√≠culos, mas n√£o t√™m funcion√°rios registados:")
df_veiculos.join(df_funcionarios, "entidade", "left_anti").show()

print("1Ô∏è‚É£4Ô∏è‚É£ Para cada entidade, calcular a m√©dia de pre√ßos dos ve√≠culos adquiridos:")
df_veiculos.groupBy("entidade").agg(avg("preco").alias("media_preco")).show()

print("1Ô∏è‚É£5Ô∏è‚É£ Criar um ranking das entidades que mais gastaram com compras de ve√≠culos:")
df_veiculos.groupBy("entidade").agg(sum(col("preco") * col("quantidade")).alias("total_gasto")).orderBy(col("total_gasto").desc()).show()

# Fechar a sess√£o
spark.stop()



```
